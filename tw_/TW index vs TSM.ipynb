```python
import yfinance as yf
import pandas
import matplotlib
```

```python
# want to find out without 2330 index performance, but how

## 1. try divide by something.

data = yf.download("^TWII 2330.TW ^TWOII",start="2020-01-01",end="2021-01-22")
data['index/tsm'] = data['Close']['2330.TW']/data['Close']['^TWII']
data['index/tsm'].plot(kind='line')
```

```python
data['normalise%_2330.TW'] = data['Close']['2330.TW']/data.iloc[0]['Close']['2330.TW']*100
data['normalise%_^TWII'] = data['Close']['^TWII']/data.iloc[0]['Close']['^TWII']*100
data['normalise%_^TWOII'] = data['Close']['^TWOII']/data.iloc[0]['Close']['^TWOII']*100
```

```python
# visualise the change of the three underlying
# it seem TWOII perform first and then tsm catch up, and not it seem TWOII fall first

data[['normalise%_2330.TW','normalise%_^TWII','normalise%_^TWOII']].plot(kind='line')
```
