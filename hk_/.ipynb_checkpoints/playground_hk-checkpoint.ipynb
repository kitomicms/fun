```python
import pandas as pd
import yfinance as yf
from bs4 import BeautifulSoup as bs
import requests
#df = yf.download('^TWII', start='2019-01-01')
html = requests.get('http://www.etnet.com.hk/www/tc/stocks/indexes_global.php')
#print(hi.text)
soup = bs(html.text,'lxml')

def func(cls):
    row_data = []
    for z in cls:
        row = soup.find_all('tr',{'class':z})
        for x in row:
            col = x.find_all('td')
            col_data = []
            for y in col:
                col_data.append(y.text)
            row_data.append(col_data)
    return row_data

cls = ['evenRow','oddRow']
result = func(cls)


header = ['ticker','最新','變動','變動率','前收市','最高','最低','香港時間','外地時間']
df = pd.DataFrame(result,columns=header)



#latest_close = 
#df['200ma'] = round((df['Adj Close'] / df['Adj Close'].rolling(200).mean() - 1 ) * 100 , 2)
```

```python
import numpy as np 
import cv2 
import pyautogui 
   
  
# take screenshot using pyautogui 
image = pyautogui.screenshot() 
   
# since the pyautogui takes as a  
# PIL(pillow) and in RGB we need to  
# convert it to numpy array and BGR  
# so we can write it to the disk 
image = cv2.cvtColor(np.array(image), 
                     cv2.COLOR_RGB2BGR) 
   
# writing it to the disk using opencv 
cv2.imwrite("image1.png", image)
```
